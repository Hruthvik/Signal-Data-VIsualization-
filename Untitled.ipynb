```python
import uproot3
import matplotlib.pyplot as plt
```

```python
file = uproot3.open('train_D02k3pi.root')
tree = file['d0tree;1']
branches = tree.arrays(namedecode = 'utf-8')
branches.keys()
```

```python
s = branches['isSignal']==1
plt.hist(branches['K_kaonID'][s],bins=100,alpha=0.5)
plt.hist(branches['K_kaonID'][~s],bins=100,alpha=0.5)
plt.title('K_kaonID')
plt.savefig("K_kaonID.jpg")
```

```python
plt.hist(branches['cosPV'][s],bins=100,alpha=0.5)
plt.hist(branches['cosPV'][~s],bins=100,alpha=0.5)
plt.title('cosPV')
plt.savefig("cosPV.jpg")
```

```python
plt.hist(branches['flightTime'][s],bins=100,range=[-0.005,0.005],alpha=0.5)
plt.hist(branches['flightTime'][~s],bins=100,range=[-0.005,0.005],alpha=0.5)
plt.title('Flight Time')
plt.savefig("flight_time.jpg")
```

```python
plt.hist(branches['chiProb'][s],bins=100,alpha=0.5)
plt.hist(branches['chiProb'][~s],bins=100,alpha=0.5)
plt.title('ChiProb')
plt.savefig("chiProb.jpg")
```

```python
plt.hist(branches['K_cosHB'][s],bins=100,alpha=0.5)
plt.hist(branches['K_cosHB'][~s],bins=100,alpha=0.5)
plt.title('K_cosHB')
plt.savefig("K_cosHB.jpg")
```
